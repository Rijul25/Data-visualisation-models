# Data visualisation challenge

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("./Documents/Dataset/movie_metadata.csv")
```

```python
df.head()
```

```python
df.columns
```

```python
titles = list(df.get("movie_title"))
```

```python
print(titles[:1])
```

```python
freq_titles = {}
for title in titles:
    length = len(title)
    
    if freq_titles.get(length) is None:
        freq_titles[length] = 1
    else:
        freq_titles[length] += 1
```

```python
freq_titles
```

```python
X = np.array(list(freq_titles.keys()))
Y = np.array(list(freq_titles.values()))
print(X)
print(Y)
```

```python
plt.scatter(X,Y)
plt.title("Movie data visualisation problem")
plt.xlabel("Length of movie title")
plt.ylabel("Freq Count")
plt.show()
```
